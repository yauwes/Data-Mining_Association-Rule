# Asosiasi Rule Behavior GYM-Member

```python
import pandas as pd 
import numpy as np
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv("gym_members_exercise_tracking.csv")
data
```

```python
data.info()
```

```python
# Fungsi untuk menghitung nilai min dan max
def min_max_values(data):
  for column in data.columns:
    unique_values = data[column].unique()
    if pd.api.types.is_numeric_dtype(data[column]):  # Cek apakah kolom bertipe numerik
      min_value = unique_values.min()
      max_value = unique_values.max()
      print(f"Minimal {column}: {min_value}")
      print(f"Maksimal {column}: {max_value}")
    else:
      print(f"Kolom {column} bukan numerik, tidak dapat dihitung min-max")

# Panggil fungsi
min_max_values(data)
```

```python
feature = ['Weight (kg)','Height (m)','Max_BPM','Avg_BPM','Resting_BPM','Calories_Burned','Fat_Percentage','Water_Intake (liters)']
data = data.drop(feature,axis=1)
```

```python
data
```

```python
def categorize_age(age):
  if age <= 25:
    return 'Sangat Muda'
  elif 26 <= age <= 35:
    return 'Muda'
  else:
    return 'Tua'

# Terapkan fungsi pada kolom Age
data['Age'] = data['Age'].apply(categorize_age)
```

```python
data
```

```python
def categorize_session(duration):
  if duration <= 0.8:
    return 'GYM-NYA Sebentar'
  elif 0.9 <= duration <= 1.3:
    return 'GYM-NYA Lumayan Lama'
  else:
    return 'GYM-NYA Lama'
# Terapkan fungsi pada kolom Durasi
data['Session_Duration (hours)'] = data['Session_Duration (hours)'].apply(categorize_session)
```

```python
data
```

```python
data['Workout_Frequency (days/week)'].unique()
```

```python
def categorize_frekuensi(frekuensi):
  if frekuensi == 2:
    return '2 Hari GYM Seminggu'
  elif frekuensi == 3:
    return '3 Hari GYM Seminggu'
  elif frekuensi == 4:
    return '4 Hari GYM Seminggu'
  else:
    return "5 Hari GYM Seminggu"
# Terapkan fungsi pada kolom Durasi
data['Workout_Frequency (days/week)'] = data['Workout_Frequency (days/week)'].apply(categorize_frekuensi)
```

```python
data
```

```python
data['Experience_Level'].unique()
```

```python
def categorize_experience(pengalaman):
  if pengalaman == 1:
    return 'Level_Pemula'
  elif pengalaman == 2:
    return 'Level_Menengah'
  else:
    return 'Level_Veteran'

data['Experience_Level'] = data['Experience_Level'].apply(categorize_experience)
```

```python
data
```

```python
def klasifikasi_bmi(bmi):
  if bmi < 18.5:
    return 'Kurus'
  elif 18.5 <= bmi <= 24.9:
    return 'Normal'
  elif 25.0 <= bmi <= 29.9:
    return 'Kelebihan Berat Badan'
  else:
    return 'Obesitas'

# Menerapkan fungsi pada kolom BMI dan membuat kolom baru
data['BMI'] = data['BMI'].apply(klasifikasi_bmi)
```

```python
# Mengubah semua kolom kategorikal menjadi bentuk one-hot encoding
data = pd.get_dummies(data)

# Pastikan data berhasil dikonversi dengan benar
data.head()
```

```python
# Set opsi tampilan pandas agar tabel tampil penuh
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.max_rows', None)     # Menampilkan semua baris (jika dibutuhkan)
pd.set_option('display.max_colwidth', None) # Menampilkan isi kolom yang panjang

# Jalankan apriori dan aturan asosiasi
frequent_itemsets = apriori(data, min_support=0.1, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0)
rules = rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])

# Tampilkan aturan asosiasi dengan kolom penuh
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]


```

```python
# Set opsi tampilan pandas agar tabel tampil penuh
pd.set_option('display.max_columns', None)   # Menampilkan semua kolom
pd.set_option('display.max_rows', None)      # Menampilkan semua baris (jika dibutuhkan)
pd.set_option('display.max_colwidth', None)  # Menampilkan isi kolom yang panjang

# Jalankan apriori dan aturan asosiasi
frequent_itemsets = apriori(data, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0)
rules = rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])

# Filter aturan yang melibatkan fitur "Workout_Type"
filtered_rules = rules[
    rules['antecedents'].astype(str).str.contains("Workout_Type") | 
    rules['consequents'].astype(str).str.contains("Workout_Type")
]

# Tampilkan aturan asosiasi yang difilter dengan kolom penuh
filtered_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
# Set opsi tampilan pandas agar tabel tampil penuh
pd.set_option('display.max_columns', None)   # Menampilkan semua kolom
pd.set_option('display.max_rows', None)      # Menampilkan semua baris (jika dibutuhkan)
pd.set_option('display.max_colwidth', None)  # Menampilkan isi kolom yang panjang

# Jalankan apriori dan aturan asosiasi
frequent_itemsets = apriori(data, min_support=0.14, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.1)
rules = rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])

# Filter aturan yang melibatkan fitur "Workout_Type"
filtered_rules = rules[
    rules['antecedents'].astype(str).str.contains("Age","Workout_Type") | 
    rules['consequents'].astype(str).str.contains("Age","Workout_Type")
]

# Tampilkan aturan asosiasi yang difilter dengan kolom penuh
filtered_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

* Support: Seberapa sering item atau itemset muncul.
* Confidence: Kemungkinan consequent muncul ketika antecedent muncul.
* Lift: Kekuatan hubungan antara antecedent dan consequent dibandingkan dengan kebetulan.
* Antecedent & Consequent: Sisi kiri dan kanan dari aturan.
 
untuk lift diatas 1 maka berkorelasi meningkat
