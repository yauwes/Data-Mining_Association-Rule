```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('online_retail_II.xlsx')
```

```python
df.shape
```

```python
df['Description'] = df['Description'].str.strip()
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['Invoice'], inplace=True)
df['Invoice'] = df['Invoice'].astype('str')
df = df[~df['Invoice'].str.contains('C')]
```

```python
basket = (df[df['Country'] =="France"].groupby(['Invoice', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Invoice'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
```

```python
# Build up the frequent items
frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)
frequent_itemsets.head()
```

```python
from mlxtend.frequent_patterns import association_rules

# Create the rules, specifying the desired number of itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1, support_only=False, num_itemsets=2) # Add num_itemsets argument
rules
```

```python
rules[ (rules['lift'] >= 5) &
       (rules['confidence'] >= 0.7) ]
```
